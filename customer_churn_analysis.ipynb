```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"Downloads/customer churn data/churn-bigml-20.csv")
df.head()
```

```python
df.groupby('Churn')[['Total day minutes', 'Total day charge']].mean()
```

```python
df.groupby('Churn')[['Total day minutes', 'Total day charge']].mean()
```

```python
pd.crosstab(df['Voice mail plan'], df['Churn'], normalize='index') * 100
```

```python
df['Tenure_Group'] = pd.cut(
    df['Account length'],
    bins=[0, 100, 200, 300, 400],
    labels=['0-100', '101-200', '201-300', '301-400']
)

pd.crosstab(df['Tenure_Group'], df['Churn'], normalize='index') * 100
```

```python
pd.crosstab(
    [df['International plan'], df['Voice mail plan']],
    df['Churn'],
    normalize='index'
) * 100
```

```python
df.to_csv("cleaned_customer_churn.csv", index=False)
```
